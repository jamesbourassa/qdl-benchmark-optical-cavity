# Optical Cavity Residual-First Benchmark

Baseline model: f = m c / (2L)

```python
import pandas as pd
import numpy as np
from statsmodels.stats.stattools import durbin_watson

# Load data directly from GitHub
url = "https://raw.githubusercontent.com/jamesbourassa/qdl-benchmark-optical-cavity/main/data.csv"
data = pd.read_csv(url)

# Quick sanity check
print("Columns:", data.columns)
print("Number of observations:", len(data))

# Constants
c = 299792458

# Extract variables
m = data["mode"].values
f = data["frequency"].values
```

```python
# Fit baseline model (estimate cavity length)
L_est = np.mean(m * c / (2 * f))

# Predictions
baseline_pred = m * c / (2 * L_est)

# Reduced model (fixed length)
L0 = 0.1
reduced_pred = m * c / (2 * L0)

# Residuals
res_baseline = f - baseline_pred
res_reduced = f - reduced_pred

# Metrics
rmse_baseline = np.sqrt(np.mean(res_baseline**2))
rmse_reduced = np.sqrt(np.mean(res_reduced**2))

dw_baseline = durbin_watson(res_baseline)
dw_reduced = durbin_watson(res_reduced)

# Output
print("\n--- Optical Cavity Benchmark ---")
print(f"Estimated cavity length L: {L_est:.6f} m")
print(f"RMSE (baseline): {rmse_baseline:.2e}")
print(f"RMSE (reduced): {rmse_reduced:.2e}")
print(f"Durbin–Watson (baseline): {dw_baseline:.3f}")
print(f"Durbin–Watson (reduced): {dw_reduced:.3f}")
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(res_baseline, label="Baseline residuals")
plt.plot(res_reduced, label="Reduced residuals")
plt.axhline(0)
plt.legend()
plt.title("Residual comparison")
plt.xlabel("Observation")
plt.ylabel("Residual")
plt.show()
```
