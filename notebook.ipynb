# Optical Cavity Residual-First Benchmark

Baseline model: f = m c / (2L)

```
import pandas as pd
import numpy as np
from statsmodels.stats.stattools import durbin_watson

c = 299792458
data = pd.read_csv('data/data.csv')

m = data['mode'].values
f = data['frequency'].values

# Fit baseline model
L_est = np.mean(m * c / (2 * f))
baseline_pred = m * c / (2 * L_est)

# Reduced model
L0 = 0.1
reduced_pred = m * c / (2 * L0)

res_baseline = f - baseline_pred
res_reduced = f - reduced_pred

rmse_baseline = np.sqrt(np.mean(res_baseline**2))
rmse_reduced = np.sqrt(np.mean(res_reduced**2))

print('Estimated L:', L_est)
print('RMSE baseline:', rmse_baseline)
print('RMSE reduced:', rmse_reduced)
print('Durbin Watson baseline:', durbin_watson(res_baseline))
print('Durbin Watson reduced:', durbin_watson(res_reduced))
```
